```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed dataset
data = pd.read_csv("../data/preprocessed_indian_ev_data.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Price
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], bins=20, kde=True)
plt.title("Distribution of EV Prices")
plt.xlabel("Price (INR)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot: Price vs Range per Charge
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Range per Charge (km)', y='Price')
plt.title("Price vs Range per Charge")
plt.xlabel("Range per Charge (km)")
plt.ylabel("Price (INR)")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
numeric_data = data.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd

# Load the clustered dataset
file_path = "C:/ML-project3-indian-ev/data/clustered_indian_ev_data.csv"  # Use the correct path
clustered_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the clustered dataset:")
print(clustered_data.head())
```

```python
# Analyze clusters using group-by operations
# Group by 'Cluster' and calculate mean values for numeric columns
cluster_summary = clustered_data.groupby('Cluster').mean()

# Display the summary statistics for each cluster
print("\nSummary Statistics for Each Cluster:")
print(cluster_summary)

# Optionally, save the cluster summary to a CSV file
cluster_summary.to_csv("data/cluster_summary.csv", index=True)
print("Cluster summary saved to 'data/cluster_summary.csv'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: Price vs Range per Charge with clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=clustered_data, x='Range per Charge (km)', y='Price', hue='Cluster', palette='Set1')
plt.title("Clusters: Price vs Range per Charge")
plt.xlabel("Range per Charge (km)")
plt.ylabel("Price (INR)")

# Save the plot
plt.savefig("reports/clusters_price_vs_range.png")  # Save the plot
plt.show()
plt.close()  # Close the plot to avoid hanging
```
